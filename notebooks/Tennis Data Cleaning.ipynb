```python
import os
import re
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("atp-200.csv")
```

```python
df.head()
```

```python
df_service = df['service-record']
```

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 500)
```

```python
df_service = df_service.apply(lambda x: json.loads(x))
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
df_service
```

```python
df_service = df_service.apply(pd.Series)
```

```python
df_service = df_service.dropna()
```

```python
df_service = df_service.reset_index()
```

```python
df_service = df_service.drop(columns = ['index'])
```

```python
df_service
```

```python
def clean_dicts(x_df):
    for col_index in range(x_df[[0]].size):
        x_df.iloc[col_index] = x_df.iloc[col_index].apply(lambda x: list(x.values())[0])
    return x_df
```

```python
clean_dicts(df_service)
```

```python
df_service = df_service.drop(0, axis = 1)
```

```python
df_service.columns = ['delete', 'Aces', 'delete2', 'Double Faults', 'delete3', '1st Serve', 'delete4', '1st Serve Points Won',
                   'delete5', '2nd Serve Points Won', 'delete6', 'Break Points Faced', 'delete7', 'Break Points Saved',
                   'delete8', 'Service Games Played', 'delete9', 'Service Games Won', 'delete10', 'Total Service Points Won']
```

```python
df_service.head()
```

```python
df_service = df_service.drop(columns = ['delete', 'delete2', 'delete3', 'delete4', 'delete5', 'delete6', 'delete7', 'delete8',
                        'delete9', 'delete10'])
```

```python
df_service
```

```python
df_service.to_csv("clean_service_stats200.csv")
```
